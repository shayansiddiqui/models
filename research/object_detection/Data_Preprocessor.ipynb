## Converting image type

```python
# For converting tiff to JPEG
def convert_to_jpg(yourpath):
    for root, dirs, files in os.walk(yourpath, topdown=False):
        for name in files:
            print(os.path.join(root, name))
            if os.path.splitext(os.path.join(root, name))[1].lower() == ".tif":
                if os.path.isfile(os.path.splitext(os.path.join(root, name))[0] + ".jpg"):
                    print("A jpeg file already exists for %s" % name)
                # If a jpeg is *NOT* present, create one from the tiff.
                else:
                    outfile = os.path.splitext(os.path.join(root, name))[0] + ".jpg"
                    try:
                        im = Image.open(os.path.join(root, name))
                        print("Generating jpeg for %s" % name)
                        im.thumbnail(im.size)
                        im.save(outfile, "JPEG", quality=100)
                    except Exception as e:
                        print(e)

convert_to_jpg('ETIS/Ground Truth/')
```

## Data Augmentation

```python
# Defining function for data augmentation
import imgaug as ia
from imgaug import augmenters as iaa
import numpy as np

# Sometimes(0.5, ...) applies the given augmenter in 50% of all cases,
# e.g. Sometimes(0.5, GaussianBlur(0.3)) would blur roughly every second image.
sometimes = lambda aug: iaa.Sometimes(0.5, aug)

# Define our sequence of augmentation steps that will be applied to every image
# All augmenters with per_channel=0.5 will sample one value _per image_
# in 50% of all cases. In all other cases they will sample new values
# _per channel_.

fliplr = iaa.Fliplr(1.0)
flipud = iaa.Flipud(1.0)
crop = iaa.CropAndPad(percent=(-0.05, 0.25))
affine = iaa.Affine(
            scale={"x": (0.8, 1.2), "y": (0.8, 1.2)}, # scale images to 80-120% of their size, individually per axis
            translate_percent={"x": (-0.2, 0.2), "y": (-0.2, 0.2)}, # translate by -20 to +20 percent (per axis)
            rotate=(-45, 45), # rotate by -45 to +45 degrees
            shear=(-16, 16), # shear by -16 to +16 degrees
            order=[0, 1], # use nearest neighbour or bilinear interpolation (fast)
            cval=(0, 0), # if mode is constant, use a cval between 0 and 255
            mode="constant" # use any of scikit-image's warping modes (see 2nd image from the top for examples)
        )

others = iaa.SomeOf((0, 5),
            [
                sometimes(iaa.Superpixels(p_replace=(0, 1.0), n_segments=(20, 200))), # convert images into their superpixel representation
                iaa.OneOf([
                    iaa.GaussianBlur((0, 3.0)), # blur images with a sigma between 0 and 3.0
                    iaa.AverageBlur(k=(2, 7)), # blur image using local means with kernel sizes between 2 and 7
                    iaa.MedianBlur(k=(3, 11)), # blur image using local medians with kernel sizes between 2 and 7
                ]),
                iaa.Sharpen(alpha=(0, 1.0), lightness=(0.75, 1.5)), # sharpen images
                iaa.Emboss(alpha=(0, 1.0), strength=(0, 2.0)), # emboss images
                # search either for all edges or for directed edges,
                # blend the result with the original image using a blobby mask
                iaa.SimplexNoiseAlpha(iaa.OneOf([
                    iaa.EdgeDetect(alpha=(0.5, 1.0)),
                    iaa.DirectedEdgeDetect(alpha=(0.5, 1.0), direction=(0.0, 1.0)),
                ])),
                iaa.AdditiveGaussianNoise(loc=0, scale=(0.0, 0.05*255), per_channel=0.5), # add gaussian noise to images
                iaa.OneOf([
                    iaa.Dropout((0.01, 0.1), per_channel=0.5), # randomly remove up to 10% of the pixels
                    iaa.CoarseDropout((0.03, 0.15), size_percent=(0.02, 0.05), per_channel=0.2),
                ]),
                iaa.Invert(0.05, per_channel=True), # invert color channels
                iaa.Add((-10, 10), per_channel=0.5), # change brightness of images (by -10 to 10 of original value)
                iaa.AddToHueAndSaturation((-20, 20)), # change hue and saturation
                # either change the brightness of the whole image (sometimes
                # per channel) or change the brightness of subareas
                iaa.OneOf([
                    iaa.Multiply((0.5, 1.5), per_channel=0.5),
                    iaa.FrequencyNoiseAlpha(
                        exponent=(-4, 0),
                        first=iaa.Multiply((0.5, 1.5), per_channel=True),
                        second=iaa.ContrastNormalization((0.5, 2.0))
                    )
                ]),
                iaa.ContrastNormalization((0.5, 2.0), per_channel=0.5), # improve or worsen the contrast
                iaa.Grayscale(alpha=(0.0, 1.0)),
#                 sometimes(iaa.ElasticTransformation(alpha=(0.5, 3.5), sigma=0.25)), # move pixels locally around (with random strengths)
#                 sometimes(iaa.PiecewiseAffine(scale=(0.01, 0.05))), # sometimes move parts of the image around
#                 sometimes(iaa.PerspectiveTransform(scale=(0.01, 0.1)))
            ],
            random_order=True
        )
```

```python
def augment_data(data_folder):
    patches = []
    masks = []
    aug_folder = 'data/train_aug'

    for dirName, subdirList, fileList in os.walk(data_folder):
        for filename in fileList:
            if (filename.endswith("jpg") or filename.endswith("jpeg")) and not filename.startswith('p'):
                print(filename)
                try:                     
                    im = Image.open(os.path.join(dirName, filename))
                    im_gt = Image.open(os.path.join(dirName, 'p'+filename))
                    
#                   left right mirror
                    fliplr_img = fliplr.augment_image(np.asarray(im))
                    fliplr_img_gt = fliplr.augment_image(np.asarray(im_gt))
                    
                    im_lr = Image.fromarray(fliplr_img)
                    im_gt_lr = Image.fromarray(fliplr_img_gt)

                    outfile = str.split(filename,'.')[0] + "_lr.jpg"
                    outfile_gt = 'p'+str.split(filename,'.')[0] + "_lr.jpg"   
                    
                    im_lr.thumbnail(im_lr.size)
                    im_lr.save(os.path.join(aug_folder, outfile), "JPEG", quality=100)
                    
                    im_gt_lr.thumbnail(im_gt_lr.size)
                    im_gt_lr.save(os.path.join(aug_folder, outfile_gt), "JPEG", quality=100)
                    
                    
#                   up down mirror
                    flipud_img = flipud.augment_image(np.asarray(im))
                    flipud_img_gt = flipud.augment_image(np.asarray(im_gt))
                    
                    im_ud = Image.fromarray(flipud_img)
                    im_gt_ud = Image.fromarray(flipud_img_gt)

                    outfile = str.split(filename,'.')[0] + "_ud.jpg"
                    outfile_gt = 'p'+str.split(filename,'.')[0] + "_ud.jpg"   
                    
                    im_ud.thumbnail(im_ud.size)
                    im_ud.save(os.path.join(aug_folder, outfile), "JPEG", quality=100)
                    
                    im_gt_ud.thumbnail(im_gt_ud.size)
                    im_gt_ud.save(os.path.join(aug_folder, outfile_gt), "JPEG", quality=100)                    
                    
#                   crop
                    crop_d = crop.to_deterministic()
                    crop_img = crop_d.augment_image(np.asarray(im))
                    crop_img_gt = crop_d.augment_image(np.asarray(im_gt))
                    
                    im_crop = Image.fromarray(crop_img)
                    im_gt_crop = Image.fromarray(crop_img_gt)

                    outfile = str.split(filename,'.')[0] + "_crop.jpg"
                    outfile_gt = 'p'+str.split(filename,'.')[0] + "_crop.jpg"   
                    
                    im_crop.thumbnail(im_crop.size)
                    im_crop.save(os.path.join(aug_folder, outfile), "JPEG", quality=100)
                    
                    im_gt_crop.thumbnail(im_gt_crop.size)
                    im_gt_crop.save(os.path.join(aug_folder, outfile_gt), "JPEG", quality=100)                     
                    
#                   affine
                    affine_d = affine.to_deterministic()
                    affine_img = affine_d.augment_image(np.asarray(im))
                    affine_img_gt = affine_d.augment_image(np.asarray(im_gt))
                    
                    im_affine = Image.fromarray(affine_img)
                    im_gt_affine = Image.fromarray(affine_img_gt)

                    outfile = str.split(filename,'.')[0] + "_affine.jpg"
                    outfile_gt = 'p'+str.split(filename,'.')[0] + "_affine.jpg"   
                    
                    im_affine.thumbnail(im_affine.size)
                    im_affine.save(os.path.join(aug_folder, outfile), "JPEG", quality=100)
                    
                    im_gt_affine.thumbnail(im_gt_affine.size)
                    im_gt_affine.save(os.path.join(aug_folder, outfile_gt), "JPEG", quality=100)
                    
#                   others
                    others_d = others.to_deterministic()
                    others_img = others_d.augment_image(np.asarray(im))
                    others_img_gt = np.asarray(im_gt)
                    
                    im_others = Image.fromarray(others_img)
                    im_gt_others = Image.fromarray(others_img_gt)

                    outfile = str.split(filename,'.')[0] + "_others.jpg"
                    outfile_gt = 'p'+str.split(filename,'.')[0] + "_others.jpg"   
                    
                    im_others.thumbnail(im_others.size)
                    im_others.save(os.path.join(aug_folder, outfile), "JPEG", quality=100)
                    
                    im_gt_others.thumbnail(im_gt_others.size)
                    im_gt_others.save(os.path.join(aug_folder, outfile_gt), "JPEG", quality=100)                    
                    
                    
                except Exception as e:
                    print(e)
#                 create_patches(os.path.join(dirName), os.path.join(dirName, filename), 25, 25, 0, patches)
#                 create_patches(os.path.join(dirName), os.path.join(dirName, 'p' + filename), 25, 25, 0, masks)
                


augment_data('data/train_aug');
# postive_patches_flipud = flipud.augment_images(postive_patches)
# postive_patches_crop = crop.augment_images(postive_patches)
# postive_patches_affine = affine.augment_images(postive_patches)
# postive_patches_others = others.augment_images(postive_patches)
```

## Generate CSV data

```python
def bbox2(img):
    a = np.where(img != 0)
    if(len(a[0])>0 and len(a[1])):
        bbox = np.min(a[1]), np.min(a[0]), np.max(a[1]), np.max(a[0])
    else:
        bbox = 0,0,0,0
    return bbox
```

```python
bb_dict={}

def find_boxes(data_folder):
    patches = []
    masks = []

    for dirName, subdirList, fileList in os.walk(data_folder):
        for filename in fileList:
            if (filename.endswith("jpg") or filename.endswith("jpeg")) and not filename.startswith('p'):
                img = Image.open(os.path.join(dirName, 'p' + filename))
                width, height = img.size
                box = bbox2(np.asarray(img))
                name, ext = filename.split('.')
                print(filename+','+str(width)+','+str(height)+',polyp'+','+str(box[0])+','+str(box[1])+','+str(box[2])+','+str(box[3]))
                bb_dict[filename]=box
```

```python
train_folder = 'data/train_aug'
test_folder = 'data/val'
print("Training data ")
print("filename,width,height,class,xmin,ymin,xmax,ymax")
find_boxes(train_folder)
print("Val data ")
print("filename,width,height,class,xmin,ymin,xmax,ymax")
find_boxes(test_folder)
```
